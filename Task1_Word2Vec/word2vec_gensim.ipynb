```python
from pathlib import Path
import pandas as pd
import jieba
from gensim.models import Word2Vec
```

```python
data_path = Path("train.csv")
if not data_path.exists():
    data_path = Path("train (1).csv")

data = pd.read_csv(data_path)
corpus = data["comment"].astype(str).tolist()

def clean_text(text):
    punctuation = [
        "，", ",", "!", "！", "。", "~", "；", ";", "？", "?",
        "【", "】", "#", "、", "（", "）", "(", ")", "“", "”",
        "：", ":", "……", "…", "*", "@", "-", "—", ".", "❤", "❤️"
    ]
    for ch in punctuation:
        text = text.replace(ch, " ")
    return text

sentences = [jieba.lcut(clean_text(text)) for text in corpus]
sentences = [[token.strip() for token in sentence if token.strip()] for sentence in sentences]

model = Word2Vec(
    sentences=sentences,
    sg=1,              # Skip-Gram
    vector_size=300,
    window=5,
    min_count=1,
    workers=1,
    seed=42
)

print("Skip-Gram Word2Vec 模型训练完成")
```
